```python
import pickle
import geopandas as gpd

# 读取 pickle 文件
with open('./hlg/init_plan_hlg.pickle', 'rb') as f:
    data = pickle.load(f)

# 提取 GeoDataFrame 对象
gdf_simplified = data['gdf']

# 保存为 GeoJSON 格式
gdf_simplified.to_file('./hlg/init_plan_hlg.geojson', driver='GeoJSON')
```

```python
import pickle
import geopandas as gpd
# 读取 GeoJSON 文件
gdf_simplified = gpd.read_file('./hlg/init_plan_hlg.geojson')

# 构造字典
d = dict()
d['gdf'] = gdf_simplified

# 将字典内容保存到 pickle 文件
with open('./hlg/init_plan_hlg.pickle', 'wb') as f:
    pickle.dump(d, f)
```
